### Text Processing 

```python
import nltk
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

## Sample
text_word = "I Love programming. "
sent_word = "I Love programming and i also love python, its my passion"

token_words1 = word_tokenize(text_word)
token_words1

token_words2 = word_tokenize(sent_word)
token_words2
```

```python
for i in token_words1:
    word_tokenize(i)
    print(i)
```

```python
print(token_words2)
```

```python
for i in token_words2:
    x = word_tokenize(i)
    print(x)
```

```python
len(token_words2)
```
