```python
import nltk

# Download required datasets
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
```

```python
# ================================
# Step 1: Import WordNetLemmatizer 
# ================================

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# ==================================
# Step 2: Lemmatize Individual Words 
# ==================================

words = [
    "running",
    "studies",
    "cars",
    "mice",
    "children",
    "better"
]

for word in words:
    print(f"{word} => {lemmatizer.lemmatize(word)} ")


# Why? running  => Remains (but running ??)
# Because the lemmatizer assumes every word is a noun unless we specify its Part-of-Speech (POS).
```

```python
# ==================================
# Step 3: Lemmatization with POS Tags
# ==================================

print(lemmatizer.lemmatize("Running", pos="v"))
print(lemmatizer.lemmatize("Better", pos="a"))
print(lemmatizer.lemmatize("am", pos="v"))
print(lemmatizer.lemmatize("went", pos="v"))
```

```python
## 
words = [
    "Goods",
    "hungry",
    "gone",
    "goes",
    "are",
    "should"
]

for word in words:
    print(lemmatizer.lemmatize(word, pos="v"))
```

## POS Codes

| POS | Meaning |
|------|---------|
| n | Noun |
| v | Verb |
| a | Adjective |
| r | Adverb |
| s | Satellite adjective |

```python
# ===================================
# Step 4: Lemmatize an Entire Sentence
# ===================================

sentence = "The boys running faster than the girls"

words = sentence.split()

lemmas = [lemmatizer.lemmatize(word) for word in words]

print("Orignal Sentence: ")
print(sentence)

print("\nAfter Lemmatization:")
print(" ".join(lemmas))
```

```python
#==================================
# Step 5: Automatic POS Tagging
#==================================

from nltk import pos_tag
from nltk.tokenize import word_tokenize

sentence = "The boys running faster than the girls"

tokens = word_tokenize(sentence)
tags = pos_tag(tokens)
tags
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(tag):

    if tag.startswith('J'):
        return wordnet.ADJ

    elif tag.startswith('V'):
        return wordnet.VERB

    elif tag.startswith('N'):
        return wordnet.NOUN

    elif tag.startswith('R'):
        return wordnet.ADV

    else:
        return wordnet.NOUN
```

```python
#==================================
# Step 6: Complete Lemmatization with POS
#==================================

tokens = word_tokenize(sentence)

tagged = pos_tag(tokens)

lemmatized_sentence = []

for word, tag in tagged:

    pos = get_wordnet_pos(tag)

    lemma = lemmatizer.lemmatize(word, pos)

    lemmatized_sentence.append(lemma)

print("Original:")
print(sentence)

print("\nLemmatized:")
print(" ".join(lemmatized_sentence))
```
