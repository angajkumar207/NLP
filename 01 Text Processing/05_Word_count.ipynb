```python
### Spacy implementation

import spacy

nlp = spacy.load("en_core_web_sm")

def count_words_spacy(text):
    doc = nlp(text)
    words = [token.text.lower() for token in doc  if not token.is_punct]
    return len(words)


text = "Let's example with python"
doc = nlp(text)
print([token.doc for token in doc])
print(count_words_spacy(text))
    
```

```python

from collections import Counter

def word_frequency(text):
    doc = nlp(text)
    words = [token.text.lower() for token in doc  if not token.is_punct]
    return Counter(words)

text = "NLP is powerful, NLP is fun"

print(word_frequency(text))
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

def categorize_sentences(text):
    sentences = sent_tokenize(text)
    categorized = {"short": [], "long": []}
    for sentence in sentences:
        doc = nlp(sentence)
        word_count = len([token.text.lower() for token in doc  if not token.is_punct])
        if word_count < 5:
            categorized["short"].append(sentence)

        else:
            categorized["long"].append(sentence)

        return categorized
```

```python
text = "Hello NLP is fun,  I like it very much . I hope everyone seeing this code"
categoriese = categorize_sentences(text)
print(categoriese)
print("Short: ", categoriese['short'])
print("Long ", categoriese["long"])
```

```python
 # Question ide..
def question_category(text):
    sentences = sent_tokenize(text)
    categorized = {"Question": [], "Not a Question": []}
    for sentence in sentences:
        if sentence.strip().endswith('?'):
            categorized["Question"].append(sentence)

        else:
            categorized["Not a Question"].append(sentence)

        return categorized
```

```python
text = "What is NLP?"
categoriese = question_category(text)
print(categoriese)
```

```python
### rules base sentiment analysis

import time

positive_words = {"happy", "good", "great", "excellent", "love", "amazing"}
negative_words = {"sad", "bad", "worse", "triable", "awful", "worst"}
```

```python
def sentiment_analysis(text):
    start_time = time.time()
    words = text.lower().split()

    pos_count = sum(1 for word in words if word in positive_words)
    neg_count = sum(1 for word in words if word in negative_words)
    sentimet = "Positive Sentiment" if pos_count > neg_count else "Negative Sentiment" if neg_count  > pos_count else "Netural Sentiment"
    end_time = time.time()
    execution_time = end_time, end_time
    return sentimet, execution_time
```

```python
# text = "I am very happy"
text = "I am very"
sentiment, exec_time = sentiment_analysis(text)
print(sentiment, exec_time)
```
