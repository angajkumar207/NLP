```python
import nltk
import spacy
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from collections import Counter

import warnings
warnings.filterwarnings("ignore")
```

```python
### Download required  resources
nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download('punkt_tab')
```

```python
### initialize the porters and stemmers

porter = PorterStemmer()
snowall = SnowballStemmer("english")
lemmatizer = WordNetLemmatizer()
nlp = spacy.load("en_core_web_sm")
```

```python
corpus = {
    "news": "The breaking developments are shifting the geopolitical landscape rapidly.",
    "wikipedia": "Photosynthesis is a process used by plants to synthesize nutrients.",
    "science_paper": "Experimental analyses demonstrated significant variations in the molecular structures."
}
```

```python
tokenize_corpus = {key: nltk.word_tokenize(value) for key, value in corpus.items()}
```

```python
print(tokenize_corpus)
```

```python
###
def apply_stemming(text_tokens, stemmer):
    return [stemmer.stem(word) for word in text_tokens]


for domain, tokens in tokenize_corpus.items():
    print(domain.upper())
    print(apply_stemming(tokens, porter))
    print(apply_stemming(tokens, snowall))
```

```python
### Applying lemmatization

def apply_lemmetization(text_tokens, lemmatizer, pos=wordnet.VERB):
    return [lemmatizer.lemmatize(word, pos) for word in text_tokens]

for domain, tokens, in tokenize_corpus.items():
    print(domain.upper())
    print(apply_lemmetization(tokens, lemmatizer))
```

```python
def spacy_lemmatization(text):
    doc = nlp(text)
    return [[token.lemma_ for token in doc]]


for domain, text in corpus.items():
    print(domain.upper())
    print(spacy_lemmatization(text))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download("punkt_tab")
nltk.download("wordnet")
```

```python
lemmatizer = WordNetLemmatizer()

GEN_Z_DICT = {
    "money": "cash", "smart": "intelligent", "excited": "thrilled", "bad": "terrible","happy": "joyful",
    "sad": "unhappy","angry": "furious","fast": "quick","slow": "sluggish","big": "large",
    "beautiful": "gorgeous","ugly": "unattractive","strong": "powerful","weak": "feeble","funny": "hilarious",
    "scared": "afraid","tired": "exhausted","hungry": "starving","rich": "wealthy","poor": "needy",
    "friend": "companion","enemy": "opponent","house": "home","car": "vehicle",
    "job": "occupation","teacher": "educator","student": "learner","kid": "child",
    "baby": "infant","food": "meal","drink": "beverage","phone": "mobile","computer": "PC",
    "book": "novel","idea": "concept","problem": "issue","answer": "solution","easy": "simple",
    "hard": "difficult","love": "adore","hate": "despise","look": "observe", "talk": "speak",
    "walk": "stroll","run": "sprint","buy": "purchase","help": "assist","fix": "repair","start": "begin","end": "finish",
}
```

```python
def normal_model(text):
    "return the text in normal from after lemmatization: "
    return " ".join(lemmatizer.lemmatize(w) for w in word_tokenize(text))


def gen_z_mode(text):
    words = word_tokenize(text.lower())
    return " ".join([GEN_Z_DICT.get(w,w) for w in words])

def translate():
    text = input("Enter a message")
    print(normal_model(text))
    print(gen_z_mode(text))
```
