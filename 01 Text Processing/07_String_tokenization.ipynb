```python
text = "Hello world, NLP is amazing. let's tokenize this sentece"

print("Word Sentence:  ", word_tokenize(text))
print("Sent word: ", sent_tokenize(text))

##3 Length
print("---------------")
print("Word Sentence: ",len(sent_tokenize(text))) 
print("Sent Sentence: ",len(word_tokenize(text)))
print("text: ",len(text))
```

```python
### Keyword extraction using TF-IDF with Tokenization

from sklearn.feature_extraction.text import TfidfVectorizer 
vectorizer = TfidfVectorizer()

x = vectorizer.fit_transform([text])

feature_array = vectorizer.get_feature_names_out()

importance = x.toarray().flatten()
importance_words = sorted(zip(feature_array, importance), key=lambda x: x[1], reverse=True)[:5]

print([word for word, _ in importance_words])
```

```python
### text Summarization

text = "Angaj helped Gunjan to go to the school in the morning"
doc = nlp(text)

words = [token.text for token in doc if token.is_alpha and token.text not in stopwords.words("english")]
filtered_freq = Counter(words)
sentence_scores = {}

for sent in doc.sents:
    for word in word_tokenize(sent.text.lower()):
        if word in filtered_freq:
            sentence_scores[sent.text] = sentence_scores.get(sent.text, 0) + filtered_freq[word]

sorted_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)
summary = " ".join(sorted_sentences[:2])
print(summary)
```

```python
from transformers import pipeline

# 1. Pipeline initialize 
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')

# 2. Text variable 
text = """hi am angaj kumar, Natural Language Processing (NLP) is a field of AI that enables
machine to understand human language, It is used Chatbots, sentiment analysis, and language translation.
"""

# 3. Model run
summary = summarizer(text, max_length=25, min_length=5, do_sample=False)
print(summary[0]['summary_text'])
```
