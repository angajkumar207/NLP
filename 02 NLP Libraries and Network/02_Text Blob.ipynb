```python
## Translator Genrator
from textblob import TextBlob
from googletrans import Translator

def  translator_blob(text, target_lang="en"):
    
    translator = Translator()
    return translator.translate(str(TextBlob(text)), dest=target_lang).text

translated = translator_blob("Hello Shubhaaaaangi", target_lang="hi")
print(translated)
```

```python
corpus = [
    "The quick brown fox easily jumped over the lazy sleeping dog.",
    "The modern artificial intelligence algorithms process unstructured data with remarkable precision.",
    "Natural language processing frameworks enable machines to comprehend and interpret human dialogue.",
    "The newly released smartphone has an absolutely stunning camera system and a phenomenal battery life, earning it a highly positive rating.",
    "While the screen resolution is excellent, the frustrating software bugs and slow performance make this device a terrible purchase.",
    "Customers provided highly subjective opinions about the restaurant's chaotic service and mediocre food.",
    "Machine learning models require massive datasets to accurately identify semantic patterns and extract meaningful information.",
    "I am incredibly excited to learn how to implement sentiment analysis using Python and the TextBlob library.",
    "Several grammatical errors and spelling mistakes were automatically corrected by the text preprocessing module.",
    "The system analyzed the conversational text to successfully extract key phrases like customer support and refund policy"
]

for i, doc in enumerate(corpus, start=1):
    blob = TextBlob(doc)
    print(i)
    print(blob.noun_phrases)
```

```python
### POST TAGING
blob = TextBlob("She is my GF and i'm loved her")

for word, tag in blob.tags:
    print(tag, word)
```

```python
### Text summarizing

from collections import Counter

text = """Python is a high-level, general-purpose programming language known for its clean syntax and readability, 
making it popular for fields like artificial intelligence, 
natural language processing (NLP), data science, and web development."""

blob = TextBlob(text)
nouns = [word.lower() for word, tag in blob.tags if tag.startswith("NN")]
common_noun = Counter(nouns).most_common(5)

Summary = " ".join([word for word, count in common_noun])
print(Summary)
```

```python
### Public  clustring

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

docs = [
    "I love python programming and Data Science",
    "Machine Learning is fascinatig",
    

]
cleaned_docs = [" ".join(TextBlob(doc).words.lower()) for doc in docs]

vectoizer = TfidfVectorizer(stop_words="english")
x = vectoizer.fit_transform(cleaned_docs)

kmeans = KMeans(n_clusters=2, random_state=42).fit(x)

for i, label in enumerate(kmeans.labels_):
    print(i, label)
```
